```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv("D:/Project/Guvi_Project/Human Voice Classification and Clustering/dataset/vocal_gender_features_new.csv")

df_cleaned = df.drop_duplicates()

#  Feature scaling (excluding 'label')
features = df_cleaned.drop("label", axis=1)
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

print(df_cleaned.dtypes.head())

# : Rebuild the DataFrame
df_final = pd.DataFrame(scaled_features, columns=features.columns)
df_final['label'] = df_cleaned['label'].values

# Save the cleaned and scaled dataset (optional)
df_final.to_csv("D:/Project/Guvi_Project/Human Voice Classification and Clustering/dataset/vocal_gender_features_new.csv", index=False)

# Summary
print("Original shape:", df.shape)
print("After cleaning shape:", df_cleaned.shape)
```
